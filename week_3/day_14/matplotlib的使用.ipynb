```python
import numpy as np
x = np.arange(1, 100)
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10), dpi=80) # 创建一个2行2列的图
ax1 = axes[0][0]
ax2 = axes[0][1]
ax3 = axes[1][0]
ax4 = axes[1][1]

ax1.plot(x,x)
ax2.plot(x,-x)
ax3.plot(x,x**2)
ax4.plot(x,np.sin(x))
plt.show()
```

```python
x = np.arange(1, 100)
fig = plt.figure(figsize=(20, 10), dpi=80) # 创建一个图
ax1 = fig.add_subplot(2,2,1) # 第一个图三个参数分别是行数，列数，第几个图
ax1.plot(x,x)
ax2 = fig.add_subplot(2,2,2)
ax2.plot(x,-x)
ax3 = fig.add_subplot(2,2,3)
ax3.plot(x,x**2)
plt.show()
```
